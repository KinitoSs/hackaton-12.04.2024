```python
df = pd.read_csv('https://github.com/KinitoSs/hackaton-12.04.2024/raw/main/result_data.csv', encoding='Windows-1251', sep=';')
df = df.drop('Unnamed: 0', axis = 1)
df
```

```python
df['docdate'] = pd.to_datetime(df['docdate'])
df['time'] = df['docdate'].dt.floor('S').dt.time
```

```python
# Преобразуем столбец 'parent' в целочисленный тип и заменяем NaN на 0
df['parent'] = df['parent'].fillna(0).astype(int)

# Заменяем все значения в столбце 'parent', которых нет в столбце 'docid', на 0
df['parent'] = df['parent'].apply(lambda x: 0 if x not in df['docid'].values else x)

# Находим все первые сообщения от пользователей, где 'parent' равен 0
user_requests = df[df['parent'] == 0][['request', 'docid']].values.tolist()

# Создаем пустой словарь для хранения цепочек диалогов
dialog_chains = {}

# Проходим по всем первым сообщениям пользователей
for request, docid in user_requests:
    # Находим все сообщения в диалоге, начиная с данного request
    dialog = df[df['docid'] == docid]
    dialog_docid = [docid]

    # Пока в диалоге есть сообщения, продолжаем искать следующие сообщения в цепочке
    while len(dialog) > 0:
        # Находим следующие сообщения в цепочке
        next_dialog = df[df['parent'].isin(dialog['docid'])]

        # Сортируем сообщения в диалоге по дате и времени
        next_dialog = next_dialog.sort_values(['date', 'hour', 'time'])

        # Проходим по каждому сообщению в следующем диалоге
        for _, row in next_dialog.iterrows():
            # Проверяем, совпадает ли реквест у текущего сообщения с исходным
            if row['request'] != request:
                # Если реквест не совпадает, заменяем его на исходный
                row['request'] = request

            # Добавляем сообщение к текущей цепочке
            dialog_docid.append(row['docid'])

        # Переходим к следующему диалогу
        dialog = next_dialog

    # Добавляем цепочку в словарь с ключом "request_docid"
    dialog_chains[f"{request}_{docid}"] = dialog_docid

# Выводим все цепочки диалогов
for request, chain in dialog_chains.items():
    print(f"Dialog request: {request}")
    print(f"Dialog chain: {chain} \n")
```

# Среднее время отклика в цепочках

```python
import datetime
import statistics

avg_response_times = []

for key, values in dialog_chains.items():
    request_messages = {}
    response_times = []
    bufer_df = df[df['docid'].isin(values)].sort_values(by='docdate', ascending=True)
    for index, row in bufer_df.iterrows():
        docdate = row['docdate']
        if row['docnumber'].startswith('О-З') or row['docnumber'].startswith('З'):
            request_messages[row['docnumber'][-4:]] = (docdate, row['docnumber'])
        elif row['docnumber'].startswith('ОП-З') and row['docnumber'][-4:] in request_messages:
            response_time = docdate - request_messages[row['docnumber'][-4:]][0]
            response_times.append(response_time)
            del request_messages[row['docnumber'][-4:]]

    if len(response_times) != 0:
        total_response_time = sum(response_times, datetime.timedelta())
        avg_response_time = total_response_time / len(response_times)
        avg_response_times.append(avg_response_time)
        print(f"Среднее время отклика поддержки для типа сообщения 'ОП' ({key}):", avg_response_time)

print("Среднее время отклика поддержки: ", (sum(avg_response_times, datetime.timedelta())/len(avg_response_times)))
median_response_time = statistics.median(avg_response_times)
print("Медианное время отклика поддержки: ", median_response_time)
```

# Время первого отклика

```python
first_response_times = []

for key, values in dialog_chains.items():
    request_messages = {}
    bufer_df = df[df['docid'].isin(values)].sort_values(by='docdate', ascending=True)
    first_response = None

    for index, row in bufer_df.iterrows():
        docdate = row['docdate']
        if row['docnumber'].startswith('О-З') or row['docnumber'].startswith('З'):
            request_messages[row['docnumber'][-4:]] = (docdate, row['docnumber'])
        elif row['docnumber'].startswith('ОП-З') and row['docnumber'][-4:] in request_messages:
            if first_response is None:
                first_response = docdate - request_messages[row['docnumber'][-4:]][0]
            del request_messages[row['docnumber'][-4:]]

    if first_response is not None:
        first_response_times.append(first_response)
        print(f"Время первого отклика поддержки для типа сообщения 'ОП' ({key}):", first_response)

if first_response_times:
    avg_first_response_time = sum(first_response_times, datetime.timedelta()) / len(first_response_times)
    print("Среднее время первого отклика: ", avg_first_response_time)
    median_first_response_time = statistics.median(first_response_times)
    print("Медианное время первого отклика: ", median_first_response_time)
else:
    print("Нет данных для вычисления среднего времени первого отклика")
```

```python
total_response_times = []

for key, values in dialog_chains.items():
    bufer_df = df[df['docid'].isin(values)].sort_values(by='docdate', ascending=True)
    first_user_message = bufer_df.iloc[0]
    last_support_message = bufer_df.iloc[-1]
    if first_user_message['docnumber'].startswith('З') and last_support_message['docnumber'].startswith('ОП-'):
        total_response_time = last_support_message['docdate'] - first_user_message['docdate']
        total_response_times.append(total_response_time)
        print(f"Общее время обработки диалога ({key}):", total_response_time)

if total_response_times:
    avg_total_response_time = sum(total_response_times, datetime.timedelta()) / len(total_response_times)
    print("Среднее общее время обработки диалога: ", avg_total_response_time)
    median_total_response_time = statistics.median(total_response_times)
    print("Медианное время обработки диалога: ", median_total_response_time)
else:
    print("Нет данных для вычисления среднего общего времени обработки диалога")
```
